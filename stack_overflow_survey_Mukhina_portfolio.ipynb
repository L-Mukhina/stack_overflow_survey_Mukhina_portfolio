```python
# Завдання №2. Аналіз повноти відповідей респондентів

# questions - створюємо множину назв питань, які одночасно присутні в qname та серед стовпців df
# використання set() та qname.unique() створює множину унікальних qname
questions = set(df1['qname'].unique())

# залишаємо тільки ті питання, які реально є у датафреймі df
questions_in_df = questions.intersection(df.columns)

# вивід повного списку питань після фільтрації
(questions_in_df)

# вивід лише перших питань для перевірки
list(questions_in_df)[:5]
```

```python
# Завдання №3. Статистичний аналіз досвіду респондентів

print(f'Міри центральної тенденції для WorkExp:')

print('Середнє:', round(df['WorkExp'].mean(), 1))
print('Медіана:', round(df['WorkExp'].median(), 1))
print('Мода:', df['WorkExp']. mode()[0])

# mean, median та mode автоматично ігнорують пропущені значення
```

```python
# Завдання №4. Аналіз віддаленої роботи

# Підраховуємо кількість рядків у df, де значення у відповідній колонці дорівнює 'Remote' або кількість респондентів, які обрали формат Remote
remote_count = df[df['RemoteWork'] == 'Remote'].shape[0]

print('Кількість респондентів, які працюють віддалено:', remote_count)
```

```python
# форма DataFrame(кількість рядків і стовпчиків)
df.shape
```

```python
# Завдання №5. Визначення популярності Python

# Створюємо новий стовпець worked_with_python, який містить True/False, де True означає, що респондент згадував Python у LanguageHaveWorkedWith
# Шукаємо слово 'Python' без урахування регістру і замінюємо пропуски (NaN) на False
df['worked_with_python'] = (df['LanguageHaveWorkedWith'].str.contains('Python', case=False, na=False))
```

```python
# Рахуємо загальну кількість респондентів, які працювали з Python
python_count = df['worked_with_python']. sum()
python_count
```

```python
# Обчислюємо частку таких респондентів у відсотках
total_respondents = df.shape[0]
python_percent = python_count / total_respondents * 100
print(f'Відсоток респондентів, які програмують на Python: {python_percent:.1f}%')
```

```python
# Завдання №6. Аналіз шляхів навчання програмуванню

# Створюємо новий стовпець Learned_with_online_courses (True/False), де True означає, що респондент навчався через онлайн-курси (використуємо стовпчик LearnCode)
df['Learned_with_online_courses'] = df['LearnCode'].str.contains('online courses', case=False, na=False)
```

```python
# Підрахуємо кількість респондентів, які навчалися онлайн
online_count = df['Learned_with_online_courses']. sum()
print(f'Кількість респондентів, які навчалися програмувати через онлайн-курси: {online_count}')
```

```python
# Завдання №7. Географічний аналіз компенсації Python-розробників

# Фільтруємо лише тих, хто працював з Python
python_devs = df[df['worked_with_python'] == True]

# Видаляємо рядки, де відсутнє значення зарплати (ConvertedCompYearly)
python_devs = python_devs.dropna(subset=['ConvertedCompYearly'])

# Групуємо за країною та обчислюємо середню і медіанну компенсацію для кожної країни
py_progers_by_country = python_devs.groupby('Country')['ConvertedCompYearly'].agg(mean_salary='mean', median_salary='median').reset_index()

# Округлюємо значення для кращої наочності
py_progers_by_country['mean_salary'] = py_progers_by_country['mean_salary'].round(1)
py_progers_by_country['median_salary'] = py_progers_by_country['median_salary'].round(1)

# Виводимо фінальний результат
py_progers_by_country
```

```python
# Завдання №8. Аналіз освіти найбільш оплачуваних спеціалістів

# Вибираємо потрібні стовпці 
top_educations = df[['ResponseId', 'EdLevel', 'ConvertedCompYearly']]

# Сортуємо за зарплатою у спадному порядку
top_educations = top_educations.sort_values(by='ConvertedCompYearly', ascending=False)

# Оновлюємо індекси після сортування, старий індекс видаляємо
top_educations = top_educations.reset_index(drop=True)

# Відбираємо перші 5 рядків з найвищою зарплатою
head5 = top_educations.head(5)

# Вивід результату (рівні освіти та зарплата)
head5[['ResponseId','EdLevel', 'ConvertedCompYearly']]
```

```python
# Завдання №9. Аналіз популярності Python по віковим категоріям

# Групуємо по вікових категоріях і обчислюємо відсотки Python-розробників
python_by_age = df.groupby("Age")["worked_with_python"].mean() * 100
print("Після групування (Series):")
print(python_by_age.head())
```

```python
# Перетворення Series у DataFrame
python_by_age = python_by_age.reset_index()
python_by_age.head()
```

```python
# Присвоєння назв стовпчиків
python_by_age.columns = ["Age", "Python Percentage"]
python_by_age.head()
```

```python
# Округлення відсотків 
python_by_age["Python Percentage"] = python_by_age["Python Percentage"]. round(1)
python_by_age.head()
```

```python
# Фінальний результат
python_by_age
```

```python
# Завдання №10. Аналіз індустрій серед високооплачуваних віддалених працівників

# Розрахуємо 75 перцентиль
comp_75 = df["ConvertedCompYearly"].quantile(0.75)
print("75-й перцентиль зарплати:", comp_75)
```

```python
# Відфільтруємо високооплачуваних віддалених працівників
high_paid_remote = df[(df["ConvertedCompYearly"] > comp_75) & (df["RemoteWork"].str.contains("Remote", case=False, na=False))]
print("Кількість високооплачуваних віддалених респондентів:", len(high_paid_remote))
```

```python
# Підрахуємо найпопулярніші індустрії серед високооплачуваних віддалених працівників і кількість респондентів
industry_counts = high_paid_remote["Industry"].value_counts().reset_index()
industry_counts.head()
```

```python
# Перейменування стовпчиків для фінального результату 
industry_counts.columns = ["Industry", "Count"]
industry_counts.head()
```
